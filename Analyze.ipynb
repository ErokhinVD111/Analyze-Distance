```python
import pandas as pd

file1 = "Антон.xlsx"
file2 = "Паровозик.xlsx"
x1 = pd.ExcelFile(file1)
x2 = pd.ExcelFile(file2)
#Sheet name: Journal
df1 = x1.parse("Journal")
df2 = x2.parse("Journal")
```

```python
df1
```

```python
df2
```

```python
time_and_coords = {}
for row1 in df1.itertuples(index=False):
    for row2 in df2.itertuples(index=False):
        if row1[0] == row2[0]:
            time_and_coords[row1[0]] = (row1[2], row1[3], row2[2], row2[3])
```

```python
list(time_and_coords.values())[:50]
```

```python
from haversine import haversine, Unit
for key, value in time_and_coords.items():
    print(f"{key}: {haversine((value[0], value[1]),(value[2], value[3]), unit=Unit.METERS)}")
```
